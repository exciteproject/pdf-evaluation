```python
import os
def size_file_dir(basedir):  
    names = os.listdir(basedir)
    paths = [os.path.join(basedir, name).replace('\\','/') for name in names]
    sizes =[]
    for path in paths:
        if path.endswith('.pdf'):
           sizes.append(int(os.stat(path).st_size/1048576))   
    return sizes
```
